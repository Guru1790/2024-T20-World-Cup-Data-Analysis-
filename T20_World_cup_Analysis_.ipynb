```python
import pandas as pd
```

```python
# Load the CSV files
batting_stats = pd.read_csv('/content/batting_stats_for_icc_mens_t20_world_cup_2024.csv')
bowling_stats = pd.read_csv('/content/bowling_stats_for_icc_mens_t20_world_cup_2024.csv')
fielding_stats = pd.read_csv('/content/fielding_stats_for_icc_mens_t20_world_cup_2024.csv')
match_results = pd.read_csv('/content/match_results_for_icc_mens_t20_world_cup_2024.csv')
wk_stats = pd.read_csv('/content/wk_stats_for_icc_mens_t20_world_cup_2024.csv')
```

```python
# Merge datasets on 'Player' column
player_stats = pd.merge(batting_stats, bowling_stats, on=['Player', 'Team', 'Mat', 'Inns'], how='outer')
player_stats = pd.merge(player_stats, fielding_stats, on=['Player', 'Team', 'Mat', 'Inns'], how='outer')
player_stats = pd.merge(player_stats, wk_stats, on=['Player', 'Team', 'Mat', 'Inns'], how='outer')
```

```python
# Display the merged data
print("Merged Player Stats:\n", player_stats.head())
```

```python
# Top run-scorers
top_run_scorers = batting_stats.nlargest(10, 'Runs')
print("Top Run Scorers:\n", top_run_scorers)
```

```python
# Top wicket-takers
top_wicket_takers = bowling_stats.nlargest(10, 'Wkts')
print("Top Wicket Takers:\n", top_wicket_takers)
```

```python
# Top fielders by catches
top_fielders = fielding_stats.nlargest(10, 'Ct')
print("Top Fielders:\n", top_fielders)
```

```python
# Top wicket-keepers by dismissals
top_wk = wk_stats.nlargest(10, 'Dis')
print("Top Wicket-Keepers:\n", top_wk)
```

```python
# Match results summary
match_summary = match_results['Winner'].value_counts()
print("Match Results Summary:\n", match_summary)
```

Visualizing the Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting top run-scorers
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='Runs', data=top_run_scorers)
plt.title('Top Run Scorers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting top wicket-takers
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='Wkts', data=top_wicket_takers)
plt.title('Top Wicket Takers')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting top fielders by catches
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='Ct', data=top_fielders)
plt.title('Top Fielders by Catches')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting top wicket-keepers by dismissals
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='Dis', data=top_wk)
plt.title('Top Wicket-Keepers by Dismissals')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting match results summary
plt.figure(figsize=(8, 5))
match_summary.plot(kind='bar')
plt.title('Match Results Summary')
plt.xlabel('Winner')
plt.ylabel('Number of Matches')
plt.show()
```

```python
# Top Batting Averages
top_batting_averages = batting_stats.nlargest(10, 'Ave')
print("Top Batting Averages:\n", top_batting_averages)
```

```python
# Plotting top batting averages
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='Ave', data=top_batting_averages)
plt.title('Top Batting Averages')
plt.xticks(rotation=45)
plt.show()
```

```python
# Best Bowling Averages
best_bowling_averages = bowling_stats[bowling_stats['Ave'] > 0].nsmallest(10, 'Ave')
print("Best Bowling Averages:\n", best_bowling_averages)
```

```python
# Plotting best bowling averages
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='Ave', data=best_bowling_averages)
plt.title('Best Bowling Averages')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top Batting Strike Rates
top_batting_sr = batting_stats.nlargest(10, 'SR')
print("Top Batting Strike Rates:\n", top_batting_sr)
```

```python
# Plotting top batting strike rates
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='SR', data=top_batting_sr)
plt.title('Top Batting Strike Rates')
plt.xticks(rotation=45)
plt.show()
```

```python
# Team performance in terms of wins
team_performance = match_results['Winner'].value_counts().reset_index()
team_performance.columns = ['Team', 'Wins']
print("Team Performance:\n", team_performance)
```

```python
# Plotting team performance
plt.figure(figsize=(10, 6))
sns.barplot(x='Team', y='Wins', data=team_performance)
plt.title('Team Performance in Terms of Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select only numeric columns for correlation analysis in batting stats
numeric_batting_stats = batting_stats.select_dtypes(include=['float64', 'int64'])
batting_corr = numeric_batting_stats.corr()
print("Batting Stats Correlation:\n", batting_corr)
```

```python
# Plotting correlation heatmap for batting stats
plt.figure(figsize=(12, 8))
sns.heatmap(batting_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Batting Stats')
plt.show()
```

```python
# Select only numeric columns for correlation analysis in bowling stats
numeric_bowling_stats = bowling_stats.select_dtypes(include=['float64', 'int64'])
bowling_corr = numeric_bowling_stats.corr()
print("Bowling Stats Correlation:\n", bowling_corr)
```

```python
# Plotting correlation heatmap for bowling stats
plt.figure(figsize=(12, 8))
sns.heatmap(bowling_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Bowling Stats')
plt.show()
```
